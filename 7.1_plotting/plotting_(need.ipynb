```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='teal', linestyle='-.')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=1)
plt.legend(loc='upper left')
# plt.show()
##Include the above as good practice, especially if this is done in a .py file


##supported line styles are: '-', '--', '-.', ':', 
##'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='teal', linestyle='-.')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.)
##alpha controls the transparency of the grid
plt.legend(loc='upper left')
plt.plot(x, y_2,
         label='Y2 values', color='blue')
##supported line styles are: '-', '--', '-.', ':', 
##'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```
